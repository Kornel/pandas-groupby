```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt

matplotlib.style.use('ggplot')
plt.rcParams["figure.figsize"] = [9, 5]

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

## Prepare data

```python
values = np.hstack((np.random.uniform(high=0.05, low=0.01, size=100),
                    np.random.uniform(high=0.06, low=0.01, size=100),
                    np.random.uniform(high=0.07, low=0.01, size=100)))

types = np.hstack((np.repeat("A", 100), np.repeat("B", 100), np.repeat("C", 100)))

df = pd.DataFrame({"values": values,
                   "types" : types})
```

## Group by and apply

```python
df.groupby("types").apply(lambda x: x["values"].plot.density());
plt.legend();
```
